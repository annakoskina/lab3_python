# **Анализ данных показывает, что Installs является текстовым полем, которое содержит разделители "," и индикатор "+" в конце строки. Необходимо преобразовать значения из строковых в числовые**

```
#Убрали последний символ в строке
q.Installs= q.Installs.map(lambda x: str(x)[:-1])
#Убрали ',' в строке
q.Installs = [x.replace(',', '') for x in q.Installs]
q.Installs
```

# **Уберем пустые значение в полях Installs и Reviews**

```
q[q.Installs=='']
q=q.dropna()#Удаляем все записи с пустыми значениями
q
```

# **Меняем тип поля с строки в float**

```
q.Installs=q.Installs.astype(float)
q.Installs
```

# **Также в результате анализа было обнаружено, что поле Size - текстовое и содержит в конце строки буквы "k", "M"(килобайты, мегабайты). Используем функцию для конвертации данных этого поля в один формат (килобайты).**

```
q.Size = q.Size.map(lambda x: float(str(x)[:-1])*1000 if x[-1] == 'M'  else (float(str(x)[:-1]) if x[-1] == 'k' else x))
q.Size
```

```

q = q[q.Size.map(lambda x: type(x) == float)]
```

```
q = q[q['Size']>=20000]
q
```

```
q_s=q #используем датафрейм q_s с нормализованными данными
q_s.Installs=q_s.Installs.transform(lambda x: (x / x.max()))
```

#**Создаем функцию для построения регрессионной модели на основе sklearn.linear_model**

```
import pickle
def model_(data_model):
   
    Y = data_model.iloc[:,[4]] # Reviews - 4 поле
    X=data_model.iloc[:,[6]]#Installs -6 поле
    
    filename = 'Model.sav'
    
    model = LinearRegression().fit(X,Y) #Обучение модели
    r_sq = model.score(X,Y) #Оценка качества модели
    print(r_sq)
    print(model.predict(X[0:1]))#Прогнозирование для 1-ого наблюдения
    pickle.dump(model, open(filename, 'wb')) #Сохранение модели на диске
```

#**Моделируем**

```
model_(q)
```

```
model_(q_s)#Это справедливо и для нормализованных данных
```

# **Функция возвращает результат в зависимости от стартовых значений поля exam1**

```
def calc_(exam):
    
    if exam>=90:
          result = 'A'

    elif exam>=80: 
          result = 'B'
          
    elif exam>=74: 
          result = 'C'
    else: 
          result = 'FF'
    return result 
```

# **Использования apply для вызова функции для каждого значения поля 'exam1'**

```
df['exam1'] = df.apply(lambda row: calc_(row['exam1']), axis=1)
df.exam1
```
